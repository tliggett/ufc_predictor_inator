```python
import pandas as pd
import numpy as np
```

```python
fight_data = pd.read_csv("data/preprocessed_data.csv")
fight_data
```

```python
fight_data = fight_data.dropna()

from sklearn.preprocessing import StandardScaler


df_num = fight_data.select_dtypes(include=[np.float, np.int])
scaler = StandardScaler()
fight_data[list(df_num.columns)] = scaler.fit_transform(fight_data[list(df_num.columns)])


fight_data
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(fight_data, test_size=0.25)
```

```python
train.to_csv('data/fight_train.csv')
test.to_csv('data/fight_test.csv')
```

```python
df = pd.read_csv("data/fight_train.csv")
fight_data = df
```

```python
def rb_binary(row):
    if row['Winner'] == 'Red':
        return 1
    return 0

fight_data['red_win'] = fight_data.apply(lambda row: rb_binary(row), axis=1)

# def southpaw_binary(row, column):
#    if row[column] == 'Southpaw':
#        return 1
#    return 0

# fight_data['B_Southpaw'] = fight_data.apply(lambda row: southpaw_binary(row, 'B_Stance'), axis=1)
# fight_data['R_Southpaw'] = fight_data.apply(lambda row: southpaw_binary(row, 'R_Stance'), axis=1)
```

```python
fight_data = fight_data.drop(columns=['Winner', 'Unnamed: 0'])
# fight_data = fight_data.drop(columns=['location', 'R_Stance', 'B_Stance', 'Referee', 'date', 'Winner', 'weight_class', 'R_fighter', 'B_fighter'])
```

```python
# Scaling
```

```python
fight_data_train, outcomes_train = fight_data.drop(columns=['red_win']), fight_data['red_win']
```

## Building an SGDClassifier
Stochastic Gradient Descent Classifier

```python
fight_data
```

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(fight_data_train, outcomes_train)
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix

outcomes_train_pred = cross_val_predict(sgd_clf, fight_data_train, outcomes_train, cv=3)
confusion_matrix(outcomes_train, outcomes_train_pred)
```

```python
from sklearn.metrics import precision_score, recall_score

```

```python
from sklearn.metrics import f1_score, roc_auc_score
print(f'F1 Score: {f1_score(outcomes_train, outcomes_train_pred)}')
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

forest_clf = RandomForestClassifier(random_state=42)

# Train the classifier
forest_clf.fit(fight_data_train, outcomes_train)

outcomes_train_pred_forest = cross_val_predict(forest_clf, fight_data_train, outcomes_train, cv=3)


print(f'Precision: {precision_score(outcomes_train, outcomes_train_pred_forest)}')
print(f'Recall: {recall_score(outcomes_train, outcomes_train_pred_forest)}')
print(f'F1 Score: {f1_score(outcomes_train, outcomes_train_pred_forest)}')
```

```python
from sklearn import metrics
```

```python
# Lets stack 'em up against the r algorithms!
f_test = pd.read_csv("data/fight_test.csv")

def rb_binary(row):
    if row['Winner'] == 'Red':
        return 1
    return 0

f_test['red_win'] = f_test.apply(lambda row: rb_binary(row), axis=1)
f_test = f_test.drop(columns=['Winner', 'Unnamed: 0'])

fight_data_test, outcomes_test = f_test.drop(columns=['red_win']), f_test['red_win']

f_test
```

```python
pred_forest = forest_clf.predict(fight_data_test)
pred_sgd = sgd_clf.predict(fight_data_test)

# Model Accuracy, how often is the classifier correct?
print("SGD Accuracy:",metrics.accuracy_score(outcomes_test, pred_sgd))
print("Forest Accuracy:",metrics.accuracy_score(outcomes_test, pred_forest))
```

## TJ Roulette Classifier

```python
class TJRouletteClassifier:
    def __init__(self):
        print('Let it ride!')
    def fit(self, x, y):
        return 0
    def predict(self, x):
        return [1] * len(x)
```

```python
roul_clf = TJRouletteClassifier()

pred_roul = roul_clf.predict(fight_data_test)
print("Roulette Accuracy:",metrics.accuracy_score(outcomes_test, pred_roul))
```

## SVM Classifier
Support Vector Machine Classifier

```python
from sklearn.svm import SVC

svm_clf = SVC()
svm_clf.fit(fight_data_train, outcomes_train)
```

```python
# Cross-Validation
outcomes_train_pred_svm = cross_val_predict(svm_clf, fight_data_train, outcomes_train, cv=3)
print(f'Precision: {precision_score(outcomes_train, outcomes_train_pred_svm)}')
print(f'Recall: {recall_score(outcomes_train, outcomes_train_pred_svm)}')
print(f'F1 Score: {f1_score(outcomes_train, outcomes_train_pred_svm)}')

# Test Case
outcomes_pred_svm = svm_clf.predict(fight_data_test)
print("SVM Accuracy:",metrics.accuracy_score(outcomes_test, outcomes_pred_svm))
```

```python
# Cross-Validation
outcomes_train_pred_knn = cross_val_predict(knn_clf, fight_data_train, outcomes_train, cv=3)
print(f'Precision: {precision_score(outcomes_train, outcomes_train_pred_knn)}')
print(f'Recall: {recall_score(outcomes_train, outcomes_train_pred_knn)}')
print(f'F1 Score: {f1_score(outcomes_train, outcomes_train_pred_knn)}')

# Test Case
outcomes_pred_knn = knn_clf.predict(fight_data_test)
print("KNN Accuracy:",metrics.accuracy_score(outcomes_test, outcomes_pred_knn))
```

## Decision Tree Classifier

```python
# Cross-Validation
outcomes_train_pred_tree = cross_val_predict(knn_clf, fight_data_train, outcomes_train, cv=3)
print(f'Precision: {precision_score(outcomes_train, outcomes_train_pred_tree)}')
print(f'Recall: {recall_score(outcomes_train, outcomes_train_pred_tree)}')
print(f'F1 Score: {f1_score(outcomes_train, outcomes_train_pred_tree)}')

# Test Case
outcomes_pred_tree = tree_clf.predict(fight_data_test)
print("Tree Accuracy:",metrics.accuracy_score(outcomes_test, outcomes_pred_tree))
```

## Voting Classifier
This is kinda dope, it takes all of the classifiers and votes to decide the outcome...

```python
from sklearn.ensemble import VotingClassifier

voting_clf = VotingClassifier(
    estimators=[('tr', tree_clf), ('svc', svm_clf), ('sgd', sgd_clf), ('fr', forest_clf), ('kn', knn_clf)],
    voting='hard'
)
voting_clf.fit(fight_data_train, outcomes_train)
```

```python
# Cross-Validation
outcomes_train_pred_vote = cross_val_predict(voting_clf, fight_data_train, outcomes_train, cv=3)
print(f'Precision: {precision_score(outcomes_train, outcomes_train_pred_vote)}')
print(f'Recall: {recall_score(outcomes_train, outcomes_train_pred_vote)}')
print(f'F1 Score: {f1_score(outcomes_train, outcomes_train_pred_vote)}')

# Test Case
outcomes_pred_vote = voting_clf.predict(fight_data_test)
print("Voting Accuracy:",metrics.accuracy_score(outcomes_test, outcomes_pred_vote))
```

## Bagging Classifier


```python
from sklearn.ensemble import BaggingClassifier

bag_clf = BaggingClassifier(
    RandomForestClassifier(random_state=42), n_estimators=500,
    max_samples=200, bootstrap=True, n_jobs=-1
)
bag_clf.fit(fight_data_train, outcomes_train)
```

```python
# Cross-Validation
outcomes_train_pred_bag = cross_val_predict(bag_clf, fight_data_train, outcomes_train, cv=3)
print(f'Precision: {precision_score(outcomes_train, outcomes_train_pred_bag)}')
print(f'Recall: {recall_score(outcomes_train, outcomes_train_pred_bag)}')
print(f'F1 Score: {f1_score(outcomes_train, outcomes_train_pred_bag)}')

# Test Case
outcomes_pred_bag = bag_clf.predict(fight_data_test)
print("Bagging Accuracy:",metrics.accuracy_score(outcomes_test, outcomes_pred_bag))
```

## Let's Get Boosting

```python
from sklearn.ensemble import AdaBoostClassifier

ada_clf = AdaBoostClassifier(
    SVC(), n_estimators=20,
    algorithm="SAMME", learning_rate=0.5
)
ada_clf.fit(fight_data_train, outcomes_train)
```

```python
# Cross-Validation
outcomes_train_pred_ada = cross_val_predict(ada_clf, fight_data_train, outcomes_train, cv=3)
print(f'Precision: {precision_score(outcomes_train, outcomes_train_pred_ada)}')
print(f'Recall: {recall_score(outcomes_train, outcomes_train_pred_ada)}')
print(f'F1 Score: {f1_score(outcomes_train, outcomes_train_pred_ada)}')

# Test Case
outcomes_pred_ada = ada_clf.predict(fight_data_test)
print("Boosting Accuracy:",metrics.accuracy_score(outcomes_test, outcomes_pred_ada))
```
